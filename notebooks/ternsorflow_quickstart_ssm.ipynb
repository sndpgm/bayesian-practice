````python
import tensorflow as tf
import tensorflow_probability as tfp

print("TensorFlow", tf.__version__)
print("TensorFlow Probability", tfp.__version__)

# 2023/04/02: MacOS M1 chip で Adam を optimizer にして ELBO 最小化ができない.
#  一方で, 以下ページに記載されているモデルは実行可能.
#  https://developer.apple.com/forums/thread/721735
# ToDo: MacOS で実行する場合には GPU ではなく CPU で実行する必要があるかどうか確認してコーディング.
tf.config.set_visible_devices([], 'GPU')
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
````

### 1.1. データ生成

````python
def create_data(t, n_observed):
    """Create sample data."""
    t_n = len(t)
    t_i = np.r_[:t_n]
    trend = 0.025 * t_i
    intercept = 10
    obs_noise = np.random.normal(0, 0.5, t_n)
    
    s = np.sin(t_i * np.pi / 3) + np.sin(t_i * np.pi / 6) + trend + intercept
    y = s + obs_noise
    
    return s, y, (t[:n_observed], y[:n_observed]), (t[n_observed:], y[n_observed:])
````

````python
t = np.arange("1996-01", "2021-01", dtype="datetime64[M]")
n_observed = int(len(t) * 0.8)
n_forecast = len(t) - n_observed
t_now = t[n_observed]

state, y, (t_observed, y_observed), (t_forecast, y_forecast) = create_data(t, n_observed)

fig, ax = plt.subplots(1, 1, figsize=(10, 4))
ax.axvline(t_now, linestyle="--")
ax.plot(t, state, lw=2, color="gray", alpha=.5, label="ground truth of state")
ax.plot(t_observed, y_observed, marker="x", markersize=3, ls="--", lw=.3, color="b", alpha=.5, label="observed for train")
ax.plot(t_forecast, y_forecast, marker="x", markersize=3, ls="--", lw=.3, color="r", alpha=.5, label="observed for test")
ax.legend()
plt.show()
````

### 1.2. モデリング・変分推論

````python
def build_model(observed_time_series):
    return tfp.sts.Sum(
        [
            tfp.sts.LocalLinearTrend(observed_time_series=observed_time_series),
            tfp.sts.Seasonal(num_seasons=12, observed_time_series=observed_time_series),
        ],
        observed_time_series=observed_time_series
    )
````

````python
model = build_model(y_observed)
````

````python
variational_posteriors = tfp.sts.build_factored_surrogate_posterior(model=model)
````

````python
num_steps = int(200)
optimizer = tf.optimizers.Adam(learning_rate=.1)

@tf.function(jit_compile=True)
def train():
    elbo_loss_curve = tfp.vi.fit_surrogate_posterior(
        target_log_prob_fn=model.joint_distribution(observed_time_series=y_observed).log_prob,
        surrogate_posterior=variational_posteriors,
        optimizer=optimizer,
        num_steps=num_steps
    )
    return elbo_loss_curve
````

````python
%%time
elbo_loss_curve = train()

fig, ax = plt.subplots(1, 1)
ax.plot(elbo_loss_curve)
plt.show()
````

### 1.3. 事後分布生成での予測

````python
ps_vi = variational_posteriors.sample(50)
````

````python
def print_params(parameters_samples):
    for name, value in parameters_samples.items():
        print(f"{name:40}: {np.mean(value, axis=0):.2f} ± {np.std(value, axis=0):.2f}")

print_params(ps_vi)
````

````python
def forecast(model, observed_time_series, parameter_samples, num_steps_forecast):
    dist = tfp.sts.forecast(
        model,
        observed_time_series=observed_time_series,
        parameter_samples=parameter_samples,
        num_steps_forecast=num_steps_forecast,
    )
    
    loc = dist.mean()[..., 0]
    scale = dist.stddev()[..., 0]
    samples = dist.sample(10)[..., 0]
    return dist, loc, scale, samples
````

````python
dist, loc, scale, samples = forecast(model, y_observed, ps_vi, n_forecast)
````

````python
def score(y, p):
    se = np.power(y - p, 2)
    mse = np.mean(se)
    r2 = np.power(np.corrcoef(y, p)[0, 1], 2)
    return mse, r2


def plot_score(y, s, p):
    def pos(co, r):
        a, b = co
        return a + (b - a) * r
    
    mse, r2 = score(y, p)
    mse_s, r2_s = score(s, p)
    print(f"       mse: {mse:7.4f}, R2: {r2:7.4f}")
    print(f" state mse: {mse_s:7.4f}, R2: {r2_s:7.4f}")
    
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax1, ax2 = ax[0], ax[1]
    
    # truth observed vs forecast
    ax1.scatter(p, y)
    ax1.set_xlabel("forecast")
    ax1.set_ylabel("ground truth observed")
    xlim = ax1.get_xlim()
    ylim = ax1.get_ylim()
    ax1.text(pos(xlim, 0.1), pos(ylim, 0.9), f'mse={mse:.4f}\n$R^2$={r2:.4f}')
    ax1.plot(xlim, ylim, c='gray')
    ax1.set_xlim(xlim)
    ax1.set_ylim(ylim)
    
    # truth state vs forecast
    ax2.scatter(p, s)
    ax2.set_xlabel("forecast")
    ax2.set_ylabel("ground truth state")
    xlim = ax2.get_xlim()
    ylim = ax2.get_ylim()
    ax2.text(pos(xlim, 0.1), pos(ylim, 0.9), f'mse={mse_s:.4f}\n$R^2$={r2_s:.4f}')
    ax2.plot(xlim, ylim, c='gray')
    ax2.set_xlim(xlim)
    ax2.set_ylim(ylim)
    

def plot_forecast_results(
    t_observed,
    y_observed,
    t_forecast,
    y_forecast,
    t,
    state,
    forecast_samples,
    forecast_mean,
    forecast_scale,
    num_step_forecast,
):
    fig, ax = plt.subplots(1, 1, figsize=(10, 5))
    ax.axvline(t_forecast[0], linestyle="--")
    ax.plot(t, state, lw=2, color="grey", alpha=.5, label="ground truth of state")
    ax.plot(t_observed, y_observed, marker="x", markersize=3, ls="--", lw=.3, color="b", alpha=.5, label="observed")
    ax.plot(t_forecast, tf.transpose(forecast_samples), ls="--", lw=.6, color="r", alpha=.1)
    ax.plot(t_forecast, forecast_mean, ls="-", lw=1., color="r", alpha=1., label="forecast")
    ax.fill_between(t_forecast, forecast_mean - 2 * forecast_scale, forecast_mean + 2 * forecast_scale, color="r", alpha=.1)
    ax.legend()
````

````python
plot_score(y[n_observed:], state[n_observed:], loc)
plt.show()

plot_forecast_results(
    t_observed,
    y_observed,
    t_forecast,
    y_forecast,
    t,
    state,
    samples,
    loc,
    scale,
    n_forecast
)
plt.show()
````

### 1.4. 成分分解

````python
def get_components(model, observed_time_series, parameter_samples, forecast_dist):
    # observed
    o = tfp.sts.decompose_by_component(
        model,
        observed_time_series=observed_time_series,
        parameter_samples=parameter_samples,
    )
    means_o = {k.name: v.mean() for k, v in o.items()}
    stddevs_o = {k.name: v.stddev() for k, v in o.items()}
    
    # forecast
    f = tfp.sts.decompose_forecast_by_component(
        model,
        forecast_dist=forecast_dist,
        parameter_samples=parameter_samples,
    )
    means_f = {k.name: v.mean() for k, v in f.items()}
    stddevs_f = {k.name: v.stddev() for k, v in f.items()}
    
    def concat_by_key(o_dict, f_dict, keys=None):
        if keys is None:
            keys = o_dict.keys()
        return {
            k: tf.concat([o_dict[k], f_dict[k]], axis=0) for k in keys
        }
    means = concat_by_key(means_o, means_f)
    stddevs = concat_by_key(stddevs_o, stddevs_f)
    return means, stddevs
````

````python
means, stddevs = get_components(model, y_observed, ps_vi, dist)
````

````python
def plot_components(t, means, stddevs, vline=None):
    N = len(means)
    fig, axes = plt.subplots(N, 1, figsize=(12, 3 * N))
    for i, name in enumerate(means.keys()):
        mean = means[name]
        stddev = stddevs[name]
        ax = axes[i]
        if vline is not None:
            ax.axvline(vline, ls="--", c="b", lw=2, alpha=.5)
        ax.plot(t, mean, lw=2)
        ax.fill_between(t, mean - 2 * stddev, mean + 2 * stddev, alpha=.2)
        ax.set_title(name)
    fig.autofmt_xdate()
    fig.tight_layout()
````

````python
plot_components(t, means, stddevs, t_now)
plt.show()
````

````python
def forecast_and_plot(
    model,
    parameter_samples,
    t_observed,
    y_observed,
    t_forecast,
    y_forecast,
    t,
    state,
    num_steps_forecast,
):
    """Summarize the above all functions."""
    print_params(parameter_samples)
    print("")
    
    dist, loc, scale, samples = forecast(model, y_observed, parameter_samples, num_steps_forecast)
    
    plot_score(y_forecast, state[len(t_observed):], loc)
    plt.show()

    plot_forecast_results(
        t_observed,
        y_observed,
        t_forecast,
        y_forecast,
        t,
        state,
        samples,
        loc,
        scale,
        num_steps_forecast
    )
    plt.show()
    
    means, stddevs = get_components(model, y_observed, parameter_samples, dist)
    plot_components(t, means, stddevs, t_forecast[0])
    plt.show()
````

````python
forecast_and_plot(
    model,
    ps_vi,
    t_observed,
    y_observed,
    t_forecast,
    y_forecast,
    t,
    state,
    n_forecast,
)
````
